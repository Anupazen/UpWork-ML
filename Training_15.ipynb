```python
import numpy as np
import pandas as pd
import torch
from torch.utils.data import DataLoader
import torch.nn as nn
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
import torch.optim as optim
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
iris = load_iris()
```

```python
print(iris.keys())
```

```python
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, shuffle=True)
```

```python
x_train = torch.FloatTensor(x_train).to(device)
x_test = torch.FloatTensor(x_test).to(device)
y_train = torch.LongTensor(y_train).to(device)
y_test = torch.LongTensor(y_test).to(device)
```

```python
class NNModel(nn.Module):
  def __init__(self,inputSize):

    super().__init__()

    self.Model = nn.Sequential(
        nn.Linear(inputSize,16),
        nn.ReLU(),

        nn.Linear(16,32),
        nn.ReLU(),

        nn.Linear(32,64),
        nn.ReLU(),

        nn.Linear(64,32),
        nn.ReLU(),

        nn.Linear(32,16),
        nn.ReLU(),

        nn.Linear(16,3),
        nn.Softmax()
    )

  def forward(self,x):
    return self.Model(x)
```

```python
nnModel = NNModel(x_train.shape[1]).to(device)
```

```python
criteria = nn.CrossEntropyLoss()
```

```python
nnOptimzer = optim.Adam(nnModel.parameters(),lr=0.001)
```

```python
epochs =100
```

```python
for epoch in range(epochs):
  op = nnModel(x_train)
  loss = criteria(op,y_train)

  nnOptimzer.zero_grad()
  loss.backward()
  nnOptimzer.step()

  print(f"epoch: {epoch+1}/{epochs}, Loss: {loss.item():.4}")
```

```python
total = 0
correct = 0

with torch.no_grad():
  op = nnModel(x_test)
  _,predicted = torch.max(op.data,1)
  total += y_test.size(0)
  correct += (predicted == y_test).sum().item()

print(f"accuracy: {(correct/total):.2%}")
```
