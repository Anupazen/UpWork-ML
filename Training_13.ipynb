```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
```

```python
iris = load_iris()
```

```python
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, shuffle=True)
```

```python
x_train = torch.tensor(x_train, dtype=torch.float32).to(device)
x_test = torch.tensor(x_test, dtype=torch.float32).to(device)
y_train = torch.tensor(y_train, dtype=torch.long).to(device)
y_test = torch.tensor(y_test, dtype=torch.long).to(device)
```

```python
class NeuralNet(nn.Module):
  def __init__(self, inputSize):
    super().__init__()

    self.Model = nn.Sequential(
      nn.Linear(inputSize, 16),
      nn.ReLU(),
      nn.Linear(16, 32),
      nn.ReLU(),
      nn.Linear(32, 16),
      nn.ReLU(),
      nn.Linear(16, 8),
      nn.ReLU(),
      nn.Linear(8, 3),
    )
  def forward(self,x):
    return self.Model(x)
```

```python
neuralNet = NeuralNet(x_train.size(1)).to(device)
```

```python
criteria = nn.CrossEntropyLoss()
optimizer = optim.Adam(neuralNet.parameters(), lr=0.005)
```

```python
epochs = 500
```

```python
totalLoss = 0.0

for epoch in range(epochs):
  output = neuralNet(x_train)

  loss = criteria(output, y_train)

  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

  totalLoss += loss.item()

  print(f'Epoch: {epoch+1}, Loss: {loss.item():.4f}')
```

```python
size = x_test.size(0)
correct = 0

with torch.no_grad():
  output = neuralNet(x_test)
  _,prediction = torch.max(output,1)
  correct += (prediction==y_test).sum().item()
print(f"accuracy : {(correct/size):.2%}")
```
